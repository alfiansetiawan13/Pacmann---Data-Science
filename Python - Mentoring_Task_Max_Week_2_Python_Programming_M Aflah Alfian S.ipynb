# Task 1. Calculate Average Sessions

**Task Descriptions**

- PacBook is a Social Media platform and they want to know what is their user Average Sessions and Session Duration for each data
- Given a three input for `load_time` (in minutes) and `exit_time` (in minutes)

- To calculate Session Duration, you can use this formula

$$
\text{Session}_\text{duration} = \text{exit}_\text{time} - \text{load}_\text{time}
$$

- To calculate Average Sessions, you can use this formula

$$
\text{Average}_\text{sessions} = \frac{\sum_{n=1}^{3}{\text{Session_duration}_{n}}}{\text{N_SESSIONS}}
$$

- But there are some constraints in this task
  - If the input for `load_time` and `exit_time` is less than zero or negative values, it will print the result **`"Not a Valid Input"`**

**Input - 1**

```py
load_time_1 = 5
exit_time_1 = 10

load_time_2 = 2
exit_time_2 = 15.5

load_time_3 = 1
exit_time_3 = 22.3
```

**Output - 1**

```py
Session 1 Duration (in minutes): 5
Session 2 Duration (in minutes): 13.5
Session 3 Duration (in minutes): 21.3
Average Session Duration (in minutes): 13.27
```

**Input - 2**

```py

load_time_2 = 0

load_time_3 = 0

```

**Output - 2**

```py

Session 3 Duration (in minutes): 0
Average Session Duration (in minutes): 0
```

**Input - 3**

```py

exit_time_2 = 5

```

**Output - 3**

```py

```

````python
# Input 1
load_time_1 = 5 #input load time 1
exit_time_1 = 10 #input exit time 1

load_time_2 = 2 #input load time 2
exit_time_2 = 15.5 #input exit time 1

load_time_3 = 1 #input load time 3
exit_time_3 = 22.3 #input exit time 3

#Process if load time and exit time >= 0
if (load_time_1 >= 0 and load_time_2 >= 0 and load_time_3 >= 0) and (exit_time_1 >= 0 and exit_time_2 >= 0 and exit_time_3 >= 0):
    session_1_duration = exit_time_1 - load_time_1 #calculate session 1 duration
    session_2_duration = exit_time_2 - load_time_2 #calculate session 2 duration
    session_3_duration = exit_time_3 - load_time_3 #calculate session 3 duration
    average = (session_1_duration + session_2_duration + session_3_duration) / 3 #calculate session average
    
    #printing each calculation
    print(f"Session 1 Duration (in minutes): {session_1_duration}")
    print(f"Session 2 Duration (in minutes): {session_2_duration}")
    print(f"Session 3 Duration (in minutes): {session_3_duration}")
    print(f"Average Session Duration (in minutes): {round(average, 2)}")
    
#if load and exit time < 0 then:
else:
    print("Not a Valid Input")
````

# Task 3. Extract NIK (Nomor Induk Kependudukan) Data

Category: Hard

---

**Task Descriptions**
- NIK is not some random numbers. You can extract information from the NIK.
  - The first six digits are a province, regency, and district code
  - The next six digits are the date of birth; DD-MM-YY, respectively
    - For men, the date of birth ranged between 01-31
    - For women, the date of birth is added with 40, hence ranged between 41-71.
  - The last 4 digits are a serial number of NIK issuance.


- Given **two** NIK, your tasks are
  1. Extract the date of birth of each NIK
  2. Extract their gender.
  3. Extract whether they have similar gender.
  4. Extract whether they have similar month of birth.

**Input**

```python
nik_1 = 1571011709860003
nik_2 = 2404075109910009
```

**Output**
```

- NIK 1 : 17-09-86
- NIK 2 : 11-09-91

Gender
- NIK 1 : Male
- NIK 2 : Female

Do they have similar gender?
No

Yes, it is on September
```

````python
# Write your code here
nik_1 = 1571011709860003 #input NIK 1
nik_2 = 2404075109910009 #input NIK 2

# Take 6 digits in NIK 1 and NIK 2 as the date of birth (DDMMYY)
dob_1 = str(nik_1)[6:12] #take DDMMYY in NIK 1
dob_2 = str(nik_2)[6:12] #take DDMMYY in NIK 2

# Take the date, month, and year
# NIK 1
date_1 = dob_1[:2] #take date NIK 1
month_1 = dob_1[2:4] #take month NIK 1
year_1 = dob_1[4:6] #take year NIK 1

# NIK 2
date_2 = dob_2[:2] #take date NIK 2
month_2 = dob_2[2:4] #take month NIK 2
year_2 = dob_2[4:6] #take year NIK 2

# Determine Gender
# What is Gender in NIK 1
if int(str(date_1)) <= 31: #if date_1 <= 31 then:
    date_1 = int(date_1) #convert date_1 from string to integer
    gender_1 = 'Male' #if date <= 31 then gender_1 is Male
else: #if date not <= 31 then:
    date_1 = int(date_1) - 40 #Convert date_1 from string to integer, then subtract it by 40 as the date result for female
    gender_1 = 'Female' #if date not <= 31 then gender_1 is Female
    
# What is Gender NIK 2
if int(date_2) <= 31: #if date_2 <= 31 then:
    date_2 = int(date_2) #convert date_2 from string to integer
    gender_2 = 'Male' #if date <= 31 then gender_2 is Male
else: #if date not <= 31 then:
    date_2 = int(date_2) - 40 #Convert date_2 from string to integer, then subtract it by 40 as the date result for female
    gender_2 = 'Female' #if date not <= 31 then gender_2 is Female

#Printing Date of Birth NIK 1 and NIK 2
print('Date of Birth') #Printing title 'Date of Birth'
print(f'- NIK 1 : {date_1}-{month_1}-{year_1}') #Printing Date of Birth NIK 1
print(f'- NIK 2 : {date_2}-{month_2}-{year_2}') #Printing Date of Birth NIK 2

#Printing Gender
print('\nGender') #Printing title 'Gender'
print(f'- NIK 1 : {gender_1}') #Printing gender type NIK 1
print(f'- NIK 2 : {gender_2}') #Printing gender type NIK 2

# Similar Gender?
if gender_1 == gender_2: #if gender 1 = gender 2 then:
    similar_gender = 'Yes' #similar gender = Yes
else: #if gendet 1 not similar gender 2 then:
    similar_gender = 'No' #similar gender = No

# Printing Similar  Gender
print('\nDo they have similar gender?') #Printing title
print(similar_gender) #Printing Result

# Similar Month?
month = {'01':'January', '02':'February', '03':'March', '04':'April', '05':'May', '06':'June', '07':'July', '08':'August',
         '09':'September', '10':'October', '11':'November', '12':'December'} 
        #Make a dict for convert from number to text of month

if int(month_1) == int(month_2): #if month 1 similar month 2 then:
    similar_month = f'Yes, it is on {month[month_1]}' #convert number to text of month
else: #if month 1 not similar month 2 then:
    similar_month = "No, it's not" 

print('\nDo they have similar month of birth?')
print(similar_month)
````

# Task 4. Calculate Pajak Penghasilan (Simplified)

Category: Hard

---

**Task Descriptions**
- According to Pemotongan Pajak Penghasilan PPh Pasal 21 Tahun 2016, we can calculate our income tax as (we'll write in Bahasa Indonesia for avoiding missinterpretation)

- **Penghasilan tidak kena pajak (PTKP)**
  - Penghasilan Rp 54.000.000 setahun untuk wajib pajak.
  - Tambahan Rp 4.500.000 diberikan jika wajib pajak menikah.
  - Tambahan Rp 4.500.000 lagi jika punya 1 orang anak atau Rp 9.000.000 jika punya 2 orang anak. Apabila jumlah tanggungan lebih dari itu, maka tetap dihitung 3 orang tanggungan.

- **Tarif Pajak Penghasilan**
  - 5% untuk penghasilan kena pajak (PKP) sampai dengan 60 Juta
  - 15% untuk PKP di atas 60 juta sampai dengan 250 Juta
  - 25% untuk PKP di atas 250 juta sampai dengan 500 juta
  - 30% untuk PKP di atas 500 juta sampai dengan 5 miliar
  - 35% untuk PKP di atas 5 miliar

- Your task is calculate the income tax given
  - Monthly income
  - Marital status
  - The number of dependents (jumlah tanggungan)

**Input - 1**

```python

marital_status = 'single'
n_dependents = 0
```

**Output - 1**

````python
# Input 1
monthly_income = 4750000 #input monthly income
marital_status = 'single' #input single or married
n_dependents = 0 #input the number of dependents 
    
#net income a year, ptkp, and pkp
if monthly_income >= 0:
    net_income_year = monthly_income*12 #hitung net income setahun
    ptkp_tk = 54_000_000 #wajib pajak ptkp setahun
    #Sudah Menikah?
    if (marital_status == 'married') or (marital_status.upper() == 'MARRIED') or (marital_status.title() == 'Married'):
        beban_menikah = 4_500_000 #pajak jika sudah menikah
        #Memiliki tanggungan ?
        if n_dependents > 0:
            beban_tanggungan = 4_500_000
            ptkp_tk =  ptkp_tk + beban_menikah + (n_dependents*beban_tanggungan) #ptkp jika memiliki tanggungan dan sudah menikah
            pkp = net_income_year - ptkp_tk #penghasilan kena pajak jika memiliki tanggungan dan sudah menikah
        else:
            ptkp_tk = ptkp_tk + beban_menikah #ptkp jika sudah menikah
            pkp = net_income_year - ptkp_tk #penghasilan kena pajak jika sudah menikah
    else:
        ptkp_tk #ptkp jika masih single
        pkp = net_income_year - ptkp_tk #penghasilan kena pajak jika masih single
else:
    print('Check the monthly_income')

#PPH 21 terhutang setahun
if pkp <= 60_000_000: #jika pkp kurang dari atau sama dengan 60 juta/tahun
    pph = pkp*0.05 #pph 5%
elif pkp <= 250_000_000: #jika pkp kurang dari atau sama dengan 250 juta/tahun
    pph = pkp*0.15 #pph 15%
elif pkp <= 500_000_000: #jika pkp kurang dari atau sama dengan 500 juta/tahun
    pph = pkp*0.25 #pph 25%
elif pkp <= 5000_000_000: #jika pkp kurang dari atau sama dengan 5 miliar/tahun
    pph = pkp*0.3 #pph 30%
else: #jika pkp di atas 5 miliar/tahun
    pph = pkp*0.35 #pph 35%
    
print(f'Penghasilan neto setahun: \n{net_income_year}')
print(f'\nPTKP TK/{n_dependents}: \n-{ptkp_tk}')
print(f'\nPenghasilan Kena Pajak: \n{pkp}')
print(f'\nPPh 21 terutang setahun: \n{round(pph)}')
````
